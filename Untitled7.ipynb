```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("jsrt_metadata.csv")
```

```python
df
```

```python
missing_data=df.isnull()
missing_data.head()
```

```python
missing_data.sum()
```

```python
sub_mean=df["subtlety"].astype('float').mean(axis=0)
sz=df["size"].astype('float').mean(axis=0)
xx=df["x"].astype('float').mean(axis=0)
yy=df["y"].astype('float').mean(axis=0)
```

```python
df["subtlety"].replace(np.nan,sub_mean,inplace=True)
df["size"].replace(np.nan,sz,inplace=True)
df["x"].replace(np.nan,xx,inplace=True)
df["y"].replace(np.nan,yy,inplace=True)

```

```python
df['position'].replace(np.nan,df['position'].value_counts().idxmax(),inplace=True)
df['diagnosis'].replace(np.nan,df['diagnosis'].value_counts().idxmax(),inplace=True)
```
